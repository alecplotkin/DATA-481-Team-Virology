```python
import pandas as pd

data = pd.read_csv('../KSV_gex_transposed.csv', index_col=0)
data
```

```python
gene_names = data.index.to_series().apply(lambda x: x.split('_')[0])
average_expression = data.groupby(gene_names).mean()
average_expression_t = average_expression.T
average_expression_t
```

```python
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

cluster_assignments = pd.read_csv('../sample_clusters.csv', index_col=0)

scaler = StandardScaler()
data_scaled = scaler.fit_transform(average_expression_t)
pca = PCA(n_components=36)
pca_result = pca.fit_transform(data_scaled)
```

```python
pca_result_df = pd.DataFrame(pca_result, index=average_expression_t.index, columns = range(1, 37))
pca_result_df['cluster'] = cluster_assignments.values
pca_result_df
```

```python
sns.scatterplot(x=pca_result_df[1], y=pca_result_df[2], hue=pca_result_df['cluster'], palette='viridis')
```

```python
pca_components = pd.DataFrame(pca.components_, index = range(1,37), columns = average_expression.index)
pca_components
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 1, figsize=(16, 12))  # Two rows, one column, larger figure size

sns.barplot(ax=axes[0], x=pca_components.loc[1].abs().sort_values(ascending=False).index, 
            y=pca_components.loc[1].abs().sort_values(ascending=False).values)
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=90)
axes[0].set_title('PCA Component 1 (Absolute Values)')

sns.barplot(ax=axes[1], x=pca_components.loc[2].sort_values(ascending=False).index, 
            y=pca_components.loc[2].sort_values(ascending=False).values, color='orange', alpha=0.6)
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=90)
axes[1].set_title('PCA Component 2 (Absolute Values)')

plt.tight_layout()
plt.show()
```

```python
top_genes_pc1 = pca_components.loc[1].abs().sort_values(ascending=False).head(10)
top_genes_pc2 = pca_components.loc[2].abs().sort_values(ascending=False).head(10)

print("Top genes contributing to PC1:")
print(top_genes_pc1)

print("\nTop genes contributing to PC2:")
print(top_genes_pc2)
```

```python
important_genes = top_genes_pc1.index.union(top_genes_pc2.index)

fig, axes = plt.subplots(5, 4, figsize=(20, 20))
axes = axes.flatten()

for i, gene in enumerate(important_genes):
    sns.boxplot(ax=axes[i], x=cluster_assignments['cluster'].values, y=average_expression_t[gene])
    axes[i].set_title(f"Expression of {gene} across clusters")
    axes[i].set_xlabel("Cluster")
    axes[i].set_ylabel("Expression Level")

plt.tight_layout()
plt.show()
```

```python
average_expression_t_cluster = average_expression_t.copy()
average_expression_t_cluster['cluster'] = cluster_assignments.values
cluster_means = average_expression_t_cluster.groupby('cluster').mean()
gene_differences = (cluster_means.iloc[0] - cluster_means.iloc[1]).abs().sort_values(ascending=False)

top_genes_differences = gene_differences.head(10)
print("Top genes with the largest differences between clusters:")
print(top_genes_differences)
```

```python
fig, axes = plt.subplots(2, 5, figsize=(20, 20))
axes = axes.flatten()

for i, gene in enumerate(top_genes_differences.index):
    sns.boxplot(ax=axes[i], x=cluster_assignments['cluster'].values, y=average_expression_t[gene])
    axes[i].set_title(f"Expression of {gene} across clusters")
    axes[i].set_xlabel("Cluster")
    axes[i].set_ylabel("Expression Level")

plt.tight_layout()
plt.show()
```

```python
# Correlation of each gene with PC1 and PC2
correlations_pc1 = average_expression_t.corrwith(pca_result_df[1]).sort_values(ascending=False)
correlations_pc2 = average_expression_t.corrwith(pca_result_df[2]).sort_values(ascending=False)

print("Top genes correlated with PC1:")
print(correlations_pc1.head(10))

print("\nTop genes correlated with PC2:")
print(correlations_pc2.head(10))
```
