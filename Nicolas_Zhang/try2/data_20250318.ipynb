```python
import pandas as pd

data = pd.read_csv('../KSV_gex_transposed.csv', index_col=0)
data
```

```python
gene_names = data.index.to_series().apply(lambda x: x.split('_')[0])
average_expression = data.groupby(gene_names).mean()
average_expression_t = average_expression.T
average_expression_t
```

```python
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

cluster_assignments = pd.read_csv('../sample_clusters.csv', index_col=0)

scaler = StandardScaler()
data_scaled = scaler.fit_transform(average_expression_t)
pca = PCA(n_components=36)
pca_result = pca.fit_transform(data_scaled)
```

```python
pca_result_df = pd.DataFrame(pca_result, index=average_expression_t.index, columns = range(1, 37))
pca_result_df['cluster'] = cluster_assignments.values
pca_result_df
```

```python
sns.scatterplot(x=pca_result_df[1], y=pca_result_df[2], hue=pca_result_df['cluster'], palette='viridis')
```

```python
pca_components = pd.DataFrame(pca.components_, index = range(1,37), columns = average_expression.index)
pca_components
```
